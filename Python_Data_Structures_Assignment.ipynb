### Q: What are data structures, and why are they important

**A:** Data structures are ways to store and organize data. They are important because they help us work with data efficiently.

### Q: Explain the difference between mutable and immutable data types with examples

**A:** Mutable data can be changed after creation, like lists.

Example:
my_list = [1, 2, 3]
my_list[0] = 10

Immutable data cannot be changed, like strings.

Example:
my_str = 'hello'
# You cannot change individual letters directly.

### Q: What are the main differences between lists and tuples in Python

**A:** Lists are mutable, meaning they can be changed. Tuples are immutable, meaning they cannot be changed.

Example:
list_example = [1, 2]
tuple_example = (1, 2)

### Q: Describe how dictionaries store data

**A:** Dictionaries store data as key-value pairs. Each key is unique and helps to quickly find its value.

### Q: Why might you use a set instead of a list in Python

**A:** Sets do not allow duplicate values. They are useful when you only need unique items.

### Q: What is a string in Python, and how is it different from a list

**A:** A string is a sequence of characters. Unlike a list, it cannot be changed directly because it is immutable.

### Q: How do tuples ensure data integrity in Python

**A:** Tuples cannot be changed, which helps protect the data from being modified by mistake.

### Q: What is a hash table, and how does it relate to dictionaries in Python

**A:** A hash table stores data using a unique key. Python dictionaries are built using hash tables.

### Q: Can lists contain different data types in Python

**A:** Yes, a list can contain different data types like numbers, strings, and even other lists.

### Q: Explain why strings are immutable in Python

**A:** Strings are immutable to make them safe to use and to help Python work faster when handling them.

### Q: What advantages do dictionaries offer over lists for certain tasks

**A:** Dictionaries are faster for searching because you use keys instead of looking through every item.

### Q: Describe a scenario where using a tuple would be preferable over a list

**A:** When you need to store fixed data like coordinates (x, y), use a tuple so the values cannot be changed.

### Q: How do sets handle duplicate values in Python

**A:** Sets automatically remove duplicate values and only keep unique items.

### Q: How does the 'in' keyword work differently for lists and dictionaries

**A:** 'in' checks for a value in a list. In a dictionary, it checks if a key exists.

### Q: Can you modify the elements of a tuple? Explain why or why not

**A:** No, you cannot modify a tuple because it is immutable.

### Q: What is a nested dictionary, and give an example of its use case

**A:** A nested dictionary is a dictionary inside another dictionary.

Example:
student = {'name': 'Ajay', 'marks': {'math': 90, 'science': 85}}

### Q: Describe the time complexity of accessing elements in a dictionary

**A:** Accessing elements in a dictionary is very fast, with a time complexity of O(1).

### Q: In what situations are lists preferred over dictionaries

**A:** Lists are preferred when order matters or when you just need to store items without unique keys.

### Q: Why are dictionaries considered unordered, and how does that affect data retrieval

**A:** Dictionaries don't store items in a fixed order. This means you can't rely on the order when retrieving data.

### Q: Explain the difference between a list and a dictionary in terms of data retrieval

**A:** Lists use positions (indexes) to retrieve data, while dictionaries use keys.
